## üìå Data visualization

#### ‚û°Ô∏è Gender distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender", hue="Gender")
plt.title("Gender distribution")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

from the above chart we have analyzed that the number of females (15424) is more than the number of males (15217)

#### ‚û°Ô∏è student scores in relation with parent education

```python
gb = df.groupby("ParentEduc").agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
```

```python
print(gb)
```

```python
plt.figure(figsize=(5,5))
plt.title("Impact of parents education on student scores")
sns.heatmap(gb, annot=True)
plt.show()
```

student whose parent education is master degree scores highiest among others and parent education with some high school is the lowest

from the above chart we have concluded that the education of parent have a good impact on students scores

#### ‚û°Ô∏è Student scores in relation with maritial status of parent

```python
gb1 = df.groupby("ParentMaritalStatus").agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
```

```python
print(gb1)
```

```python
plt.figure(figsize=(5,5))
plt.title("Impact of parents maritial status on student scores")
sns.heatmap(gb1, annot=True)
plt.show()
```

from the above chart we have concluded that there is negligible impact on students score due to their parent maritial status

#### ‚û°Ô∏è Distribution of ethinic groups

```python
print(df["EthnicGroup"].unique());
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

```

```python

l = ['Group A', 'Group B', 'Group C', 'Group D', 'Group E']
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct="%1.2f%%", wedgeprops=dict(width=0.2))
plt.title("Distribution of ethnic groups")
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
ax = sns.countplot(data = df, x = "EthnicGroup", hue="EthnicGroup")
for container in ax.containers:
    ax.bar_label(container)
plt.legend().remove()
plt.show()
```

from the above, we can conclude that Group C have most number of students that is 9212 (31.99%)
